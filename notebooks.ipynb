```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ethiopia_economic_data.csv")
df
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["GDP_Growth_%"], label="GDP Growth")
plt.plot(df["Year"], df["Inflation_%"], label="Inflation")
plt.plot(df["Year"], df["Unemployment_%"], label="Unemployment")
plt.title("Ethiopia: Key Economic Indicators (20102023)")
plt.xlabel("Year")
plt.ylabel("Percent")
plt.legend()
plt.grid(True)
plt.show()
```

## ### Key Observations

- Ethiopia experienced **high GDP growth** during early 2010s.
- **Inflation increased sharply after 2018**, indicating macroeconomic instability.
- Unemployment remained **relatively low but slowly increasing**.
- Strong negative relationship between inflation and growth after 2019.
